# Naive Bayes Classifier

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from wordcloud import WordCloud

from nltk import download
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (15, 9)
```



## Preprocessing

```python
def process_message(message):
    message = message.lower()
    words = word_tokenize(message)
    words = [w for w in words if len(w) > 2]
    sw = stopwords.words('english')
    words = [w for w in words if w not in sw]
    stem = PorterStemmer()
    words = [stem.stem(w) for w in words]
    return words
```

## Evaluation

```python
def visualize(corpus, label=1):
    mails = corpus['mail']
    labels = corpus['label']
    chosen_mails = mails[labels == label]
    words = " ".join(w for mail in chosen_mails for w in mail)
    words_wc = WordCloud().generate(words)
    plt.imshow(words_wc)
    plt.axis('off')
    plt.show()

    
def metrics(labels, predictions):
    true_pos, true_neg, false_pos, false_neg = 0, 0, 0, 0
    for i in range(len(labels)):
        true_pos += int(labels[i] == 1 and predictions[i] == 1)
        true_neg += int(labels[i] == 0 and predictions[i] == 0)
        false_pos += int(labels[i] == 0 and predictions[i] == 1)
        false_neg += int(labels[i] == 1 and predictions[i] == 0)
    precision = true_pos / np.float(true_pos + false_pos)
    recall = true_pos / np.float(true_pos + false_neg)
    Fscore = 2 * precision * recall / np.float(precision + recall)
    accuracy = (true_pos + true_neg) / np.float(true_pos + true_neg + false_pos + false_neg)

    print("Precision: ", precision)
    print("Recall: ", recall)
    print("F-score: ", Fscore)
    print("Accuracy: ", accuracy)
```

```python
class SpamClassifier():

    def __init__(self):
        self.wf_spam = {}
        self.wf_ham = {}
        self.wf = {}
        self.prob_spam = {}
        self.prob_ham = {}
        self.C_spam = 0.5
        self.C_ham = 0.5
        self.spam_words = 0
        self.ham_words = 0

    def train(self, X, y):
        
        self.no_of_messages = len(X)
        self.no_ham_messages = (y==0).sum()
        self.no_spam_messages = y.sum()
        self.calc_word_freq(X,y)
        self.calc_prob()

    def calc_word_freq(self, X, y):
        for mail, label in zip(X,y):
            for word in mail:
                if label == 1:
                    self.wf_spam[word] = self.wf_spam.get(word, 0) + 1
                    self.spam_words += 1
                else:
                    self.wf_ham[word] = self.wf_ham.get(word, 0) + 1
                    self.ham_words += 1
                self.wf[word] = self.wf.get(word, 0) + 1
        self.no_of_words = len(self.wf)

    def calc_prob(self):
        # YOUR CODE HERE
        # SPAM
        for word in self.wf_spam:
            self.prob_spam[word] = (self.wf_spam[word]) / float(self.spam_words)
        
        self.C_spam = self.no_spam_messages/float(self.no_of_messages)
        
        #NOT SPAM
        for word in self.wf_ham:
            self.prob_ham[word] = (self.wf_ham[word]) / float(self.ham_words)

        self.C_ham = self.no_ham_messages/float(self.no_of_messages)
        # END
        
    def classify(self, X):
        # YOUR CODE HERE
        list_classified = []

        for message in X:
            prob_spam_ = np.log(self.C_spam)
            prob_ham_ = np.log(self.C_ham)
            for word in message:
                try:
                    prob_spam_ += np.log(self.prob_spam[word])
                    prob_ham_ += np.log(self.prob_ham[word])
                except:
                    prob_spam_ += 1/float(1+self.no_of_words)
                    prob_ham_ += 1/float(1+self.no_of_words)

            #print("prob_spam:",np.exp(prob_spam_),"prob_ham:", np.exp(prob_ham_))
            #print(p_C_spam_wi_wn)
            #print(p_C_ham_wi_wn)
            if(np.exp(prob_spam_) > np.exp(prob_ham_)):
                list_classified.append(1)
            else:
                list_classified.append(0)
            
        return np.array(list_classified) 
        # END

        
        
```

## download data set

NLTK is a standard package for dealing with natural language data. It also contains useful data sets, such as stop words. In the following cell, some NLTK data is downloaded and the spam mail data set is loaded from file.

```python
download('punkt')
download("stopwords")

raw_in = pd.read_csv("spam.csv", encoding='latin-1')
raw_in["label"] = raw_in.v1.apply(lambda x: 0 if x == "ham" else 1)
raw_in["mail"] = raw_in.v2.apply(process_message)

corpus = pd.DataFrame(raw_in[["mail", "label"]])
corpus.head()
```

```python
print("HAM WORD CLOUD")
visualize(corpus, label=0)
print("SPAM WORD CLOUD")
visualize(corpus, label=1)
```
